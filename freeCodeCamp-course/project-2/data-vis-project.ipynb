```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_population= pd.read_csv('population_total.csv')
```

# Making a Pivot Table

```python
df_population.dropna(inplace=True)
```

```python
df_pivot=df_population.pivot(index="year", columns="country", values="population")
```

```python
df_pivot=df_pivot[['United States','Brazil','Indonesia']]
```

# Lineplot

```python
df_pivot.plot(kind='line',title='Population ( 1955 - 2020) ',xlabel='Year',ylabel='Population',figsize=(10,5))


plt.savefig('population.png')

plt.show()

```

# Barplot

```python
df_pivot_2020= df_pivot[df_pivot.index.isin([2020])]
```

```python
# transpose dataframe from rows to columns or viceversa
df_pivot_2020 = df_pivot_2020.T
```

```python
# making barplot
df_pivot_2020.plot(kind='bar', color='green',
                   xlabel='Years', ylabel='Population',
                   title='Population in 2020',)
```

##  Barplot grouped by 'n' variables

```python
# selecting a few years
df_pivot_sample = df_pivot[df_pivot.index.isin([1980, 1990, 2000, 2010, 2020])]
```

```python
# making grouped barplot
df_pivot_sample.plot(kind='bar',
                     xlabel='Years', ylabel='Population',
                     figsize=(8,4))
```

# Piechart

```python
df_pivot_2020.rename(columns={2020:'2020'},inplace=True)
```

```python
df_pivot_2020.plot(kind='pie',
                   y='2020',
                   title='Population in 2020',)
```

# Boxplots

```python
df_pivot['United States'].plot(kind='box',color='red',ylabel='Population',)
```

```python
df_pivot.plot(kind='box',color='blue',ylabel='Population',)
```

# Histogram

```python
df_pivot[['United States', 'Indonesia']].plot(kind='hist',
                                        xlabel='Population')
```
